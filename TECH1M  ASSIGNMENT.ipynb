```python
# i imported the useful or neccessary libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
```

```python
# read the train dataset(importing the dataset)


train = pd.read_csv(r'C:\Users\USER\Downloads\train.csv')
```

```python
#view the train dataset


train.head()
```

```python
#basic information about the train dataset

train.info()
```

```python
# Changing the data type of the column 'Date' from object to  datetime


train['Date'] = train['Date'].astype('Datetime64')
```

```python
# Confirming if the chnage has been effected 


train.info()
```

```python
#description of the train dataset

train.describe()
```

```python
#Find null values

train.isnull().sum()
```

```python
# Checking for missing data


for col in train.columns:
    pct_missing = np.mean(train[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Dropping all the rows with missing data

train = train.dropna()
```

```python
#Find the duplicates : Well, the function returned ‘0’. This means, there is not a single duplicate value present in our dataset and it is a very good thing to know.



train.duplicated().sum()
```

```python
# i splitted the DATE column into "year", "month", and " day"



train['month']=train['Date'].dt.month


train['year']=train['Date'].dt.year


train['day']=train['Date'].dt.day
```

```python
#view the train dataset to confirm if the change has been effected.


train
```

```python
# i created a pivot table and i added the aggregate function 'sum' to get the total number of customers, sales and the total number of days of promo in all the three years.


train.pivot =  train.pivot_table(index = "year", values = ["Promo", "Customers", "Sales"], aggfunc = "sum") 
```
